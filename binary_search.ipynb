```python
value = 891
```

```python
def binarysearch(data, value):
    beg = 0
    end = len(data) - 1

    while beg <= end:
        mid = (beg + end) // 2   # integer division
        if value == data[mid]:
            print(f'Found it! Index: {mid}')
            return  # stop once found
        elif value > data[mid]:
            beg = mid + 1
        else:
            end = mid - 1
    print("Not found.")
```

```python
binarysearch(data,value)
```
