```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
# Creating nodes
root = Node(1)
root.left = Node(2)
root.right = Node(3)

root.left.left = Node(4)
root.left.right = Node(5)
root.left.right.left = Node(8)
```

```python
print(root.left.right.left.data)
```

- print the data in PREORDER 

```python
def preOrderTraversal(node):
    if node is None:
        return
    print(node.data, end=", ")
    preOrderTraversal(node.left)
    preOrderTraversal(node.right)
```

```python
preOrderTraversal(root)
```

```python
def Postorder(node):
    if node is None:
        return
    Postorder(node.left)
    Postorder(node.right)
    print(node.data,end= ' ')
```

```python
Postorder(root)
```

```python
def inorder(node):
    if node is None:
        return
    inorder(node.left)
    print(node.data,end=' ')
    inorder(node.right)
```

```python
inorder(root)
```
