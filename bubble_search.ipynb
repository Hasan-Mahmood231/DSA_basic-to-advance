# Bubble sort:

```python
import array as arr
```

```python
num = arr.array('i',[233,3,943,11,21,2,90,1,10])
```

```python
n = len(num)
for j in range(0,n-1):
    ptr = 0
    for i in range(0,n-j-1):
        if num[ptr] > num[ptr+1]:
            tem = num[ptr]
            num[ptr] = num[ptr+1]
            num[ptr+1] = tem
        ptr+=1
        
```
